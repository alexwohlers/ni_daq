```python
##############################################
# Hardware
#    - Windows 11 
#    - NI USB-6008
#    Erst NI-DAQmx 19.5 installiert 
#          (aktuelle Version 2023 Q2 zeigt die Karte nicht an)
#          (Die Unterstützung für Net Framework und C++ nicht installieren wenn kein Compiler auf dem Rechner)
#    Dann nidaqmax installieren: "pip install nidaqmx" in Anaconda Prompt

import time as t
import numpy as np
import nidaqmx
from nidaqmx.constants import AcquisitionType
import pprint

fSampling = 10000 #Hz
nSamples = 1000 #samples
pp = pprint.PrettyPrinter(indent=4)
dtMax = 10 #sec

###############################################
# DAQ

data_A_1 = []
data_B_1 = []
t_insgesamt = 0

# config
with nidaqmx.Task() as taskA, nidaqmx.Task() as taskB:
    t_vorher = t.time()    
    taskA.ai_channels.add_ai_voltage_chan("Karte_A/ai0")    
    taskB.ai_channels.add_ai_voltage_chan("Karte_B/ai0")    
    taskA.timing.cfg_samp_clk_timing(fSampling, sample_mode=AcquisitionType.CONTINUOUS)
    taskA.timing.cfg_samp_clk_timing(fSampling, sample_mode=AcquisitionType.CONTINUOUS)

    t_nachher = t.time()    
    while t_insgesamt < dtMax:
        value_A_1 = taskA.read(number_of_samples_per_channel=nSamples)
        value_B_1 = taskB.read(number_of_samples_per_channel=nSamples)                
        #t.sleep(1)
        data_A_1.extend(value_A_1)
        data_B_1.extend(value_B_1)
        t_nachher = t.time()
        dt_buffer = t_nachher-t_vorher
        t_insgesamt += dt_buffer
        t_vorher = t_nachher
        #Verzögerungen haben keinen Einfluss auf die Messung! Super, hier der Beweis:        
        #print(dt_buffer)
        #pp.pprint(data)
###############################################
# output to file

```

```python
print("Gesamte Messzeit [s]: " + str(t_insgesamt))
print("Anzahl aufgenommener Werte Kanal A1: " + str(len(data_A_1)))
print("Anzahl aufgenommener Werte Kanal B1: " + str(len(data_B_1)))
print("Abtastfrequenz Kanal A1 [Hz]: " + str(len(data_A_1) / t_insgesamt))
print("Abtastfrequenz Kanal B1 [Hz]: " + str(len(data_B_1) / t_insgesamt))
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300

t_A = np.arange(0,len(data_A_1)/fSampling,1/fSampling)
t_B = np.arange(nSamples/fSampling,(len(data_A_1)+nSamples)/fSampling,1/fSampling)
fig = plt.figure()
plt.plot(t_A,data_A_1)
plt.plot(t_B,data_B_1)
#plt.xlim(0,1)
plt.tight_layout()
plt.xlabel('Zeit [s]')
plt.ylabel('Spannung [V]')
plt.show()
fig.savefig('..\output_bilder\output_voltage.png', dpi=fig.dpi)
```

```python
f = open("..\output_csv\output_voltage.txt", "w")
f.write("t,U")
for i in range(len(data_A_1)):
    f.write(str(i/len(data_A_1)*t_insgesamt) + ',' + str(data_A_1[i]) + '\n')
f.close()
```
